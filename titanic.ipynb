```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head() 
# opens titanic.csv to read
```

```python
data1 = df.groupby(["Pclass","Sex"])['Survived'].count()
# sets data to each columns info
```

```python

import matplotlib.pyplot as plt
data1.plot.bar(x='Sex', y = 'Pclass',rot=0);
#plots data on graph
```
