Using your knowledge of Python and Matplotlib write your code to transform and graph the data below.

After your code create a new Markdown block to describe your analysis

````python
import matplotlib.pyplot as plt

states = open('states.txt')
WA=0
MO=0
NY=0
CA=0
IL=0

for items in states.readlines():
    if items == 'WA\n':
        WA +=1
    if items == 'MO\n':
        MO +=1
    if items == 'NY\n':
        NY +=1
    if items == 'CA\n':
        CA +=1
    if items == 'IL\n':
        IL +=1
print()

labels= 'Wa','MO','NY','CA','IL'
sizes= [WA,MO,NY,CA,IL]
fig, ax =plt.subplots()
ax.pie(sizes, labels=labels)
plt.title('which is better from a unanmous vote')
````
