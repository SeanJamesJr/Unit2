Using your knowledge of Python and Matplotlib write your code to transform and graph the data below.

After your code create a new Markdown block to describe your analysis

````python
import matplotlib.pyplot as plt

states = open('states.txt')
AL=0
AK=0
AZ=0
AR=0
AS=0

for items in states.readlines():
    if items == 'WA\n':
        AL +=1
    if items == 'MO\n':
        AK +=1
    if items == 'NY\n':
        AZ +=1
    if items == 'CA\n':
        AR +=1
    if items == 'IL\n':
        AS +=1
print()

labels='AL','AK','AZ','AR','AS'
sizes= [AL,AK,AZ,AR,AS]
fig, ax =plt.subplots()
ax.pie(sizes, labels=labels)
plt.title('which is better from a unanmous vote')
````

I noticed that the more the state is mentipned in states.txt the larger it is on the pie graph
